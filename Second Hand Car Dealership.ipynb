**Data Cleaning**

```python
# Sorting the data in alphabetical order
data = data.sort_values('Name')

# Dropping New_Price as it contains majority of Null values
data = data.drop(columns = ["Unnamed: 0", 'New_Price'])
```

Data Imputing to remove null values

```python
#data["Price"].fillna(data["Price"].mean(), inplace = True)

data["Engine"].fillna(data["Engine"].mode(), inplace = True)
#data["Power"].fillna(data["Power"].mode(), inplace = True)
data["Seats"].fillna(data["Seats"].median().astype(int), inplace = True)

# Removing Null values
data = data.dropna()
```

Resetting the data index after remove unnecessary data

```python
data = data.reset_index()
```

```python
# No of unique values in Name category of the data
data["Name"].value_counts()
```

While trying to split the string data to get the float value, it was noticed that some of the data in the mileage are of km/kg while majority were of kmpl unit. Hence converting Km/kg to kmpl in Mileage and float type.

```python
Correct_Mileage= []
for i in data.Mileage:
    if str(i).endswith('km/kg'):
        i = i[:-6]
        i = float(i)*1.40
        Correct_Mileage.append(float(i))
    elif str(i).endswith('kmpl'):
        i = i[:-6]
        #print(i)
        Correct_Mileage.append(float(i))
```

```python
data['Mileage']=Correct_Mileage
```

```python
#Splitting the data Name Category to get only the brands of the cars.
data['Brand'] = data['Name'].str.split(' ').str[0]
```

While Splitting Power data, it was noticed that it contains "null" in the data even after removing NaN vales, hence Splitting it and replacing null with 0 and then with the mean of the Power data. 

```python
#data['Power'] = data['Power'].str.split(' ').str[0].astype(float)
data['Power'] =data['Power'].apply(lambda x: float(x.split()[0].replace('null', "0")))
data['Power']=data['Power'].replace(0,data['Power'].mean())
```

```python
# Splitting and converting the Engine data to get only float type values.
data['Engine'] = data['Engine'].str.split(' ').str[0].astype(float)
```

**LINEAR REGRESSION MODEL**

**Splitting the data to obtain test and train**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

From the Above correlation data,  it is found that year, Engine type, Power have positive relation on 

```python
y = data["Price"].values.reshape(-1,1) # converting it to 2D array
```

```python
x = data["Year"].values.reshape(-1,1) # converting it to 2D array
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3, random_state = 30)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
# Validating the model
def c(slope, intercept, year):
    return slope*year+intercept
price = c(reg.coef_,reg.intercept_, 2016)    
```

```python
price
```

```python
y_pred = reg.predict(x_test)
```

```python
print("Mean Squared Error is :", mean_squared_error(y_test, y_pred))
print("Mean Absolute Error is :",mean_absolute_error(y_test, y_pred))
print("Root Mean Squared  Error is :", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 Score is  :", r2_score(y_test, y_pred))
```

**Multiple LINEAR Regression**

```python
X = data[["Year","Mileage","Engine","Power"]].values
```

```python
Y = data["Price"].values.reshape(-1,1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.3, random_state = 30)
```

```python
Linreg = LinearRegression()
```

```python
Linreg.fit(X_train, Y_train)
```

```python
Linreg.intercept_
```

```python
Linreg.coef_
```

```python
Y_pred = Linreg.predict(X_test)
print(Y_pred)
```

```python
print("Mean Squared Error is :", mean_squared_error(Y_test, Y_pred))
print("Mean Absolute Error is :",mean_absolute_error(Y_test, Y_pred))
print("Root Mean Squared  Error is :",np.sqrt(mean_squared_error(Y_test, Y_pred)))
print("R2 Score is  :", r2_score(Y_test, Y_pred))
print("Residual Sum of Squares : ", (Y_test-Y_pred)**2)
```

**Validating the model**

```python
data_pred = pd.DataFrame({"Actual":Y_test.squeeze(), "Predicted":Y_pred.squeeze()})
print(data_pred)
```

The Multiple linear regression model shows the accuracy of 67.76%
